```python
import numpy as np
import pandas as pd
import seaborn as sns
from collections import OrderedDict
```

```python
npts = 6
factor = 0.1

data = []
for i in range(50):
    data.append(np.arange(npts) + np.random.randn(npts) * factor)

df = pd.DataFrame(data)
df
```

```python
ncolors = 6

for pname in ['tab10', 'colorblind', 'deep', 'bright', 'dark', 'pastel', 'bright', 'muted', 'husl']:
    # tab10 is the default palette
    print(f'palette: {pname}')
    palette = sns.color_palette(pname, ncolors)
    display(palette)
    # sns.set_palette(palette)
    df.plot(color=palette, title=pname)
```

```python
import seaborn as sns
import colorsys

tab10 = sns.color_palette('tab10')
tab10_desat = sns.color_palette('tab10', desat=0.9)

h, l, s = colorsys.rgb_to_hls(*tab10[0])

s *= 0.9

rgb = colorsys.hls_to_rgb(h, l, s)

print(tab10[0])
print(tab10_desat[0])
rgb
```
